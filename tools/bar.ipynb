```
import gzip
import requests
import brotli

def check_comp(url):
  'return (content-type, gzip_size, brotli_size, gzip_srvd, br_srvd)'

  r2 = requests.get(url)
  #print('original: {}'.format( len(r2.content) ) )

  comp = gzip.compress(r2.content, 5)
  #print('gzip: {}'.format( len(comp) ) )

  comp2 = brotli.compress(r2.content, quality=11)
  #print('brotli: {}'.format( len(comp2) ) )

  gzip_srvd = 0
  if 'content-encoding' in r2.headers and r2.headers['content-encoding'] == 'gzip':
    gzip_srvd = 1

  br_srvd = 0
  headers = {'Accept-Encoding': 'gzip,br'}
  r3 = requests.get(url, headers=headers)
  if 'content-encoding' in r3.headers and r3.headers['content-encoding'] == 'br':
    br_srvd = 1
  

  return (r2.headers['content-type'], len(r2.content), len(comp), len(comp2), gzip_srvd, br_srvd)

#print( check_comp('https://space.ktmrmshk.com/demosite/basic/bigsvg2.svg') )
#print( check_comp('https://www.misawa.co.jp/common_d/css/reset.css') )
print( check_comp('https://github.com/ktmrmshk/fract/blob/master/src/frase.py') )
```

```
from html.parser import HTMLParser
from urllib.parse import urlparse, urljoin
import requests

class BParser(HTMLParser):
  def start(self, url):
    self.url = url
    self.suburls = list()

    r = requests.get(url)
    self.feed(r.text)

  def handle_starttag(self, tag, attrs):
    if tag == 'script':
      for a in attrs:
        if a[0] == 'src':
          #print(urljoin(url,a[1]))
          self.suburls.append( urljoin(self.url,a[1]) )

    if tag == 'img':
      for a in attrs:
        if a[0] == 'src':
          #print(urljoin(url,a[1]))
          self.suburls.append( urljoin(self.url,a[1]) )

    if tag == 'link': 
      isStyleSheet = False
      href=None
      for a in attrs:
        if a[0] == 'href':
          href=a[1]
        if a[0] == 'rel':
          if a[1] == 'stylesheet':
            isStyleSheet = True
      
      if isStyleSheet:
        #print(urljoin(url,href))  
        self.suburls.append(  urljoin(self.url,href) )



def suburls(src):
  p = BParser()
  p.start(src)
  return p.suburls

# suburls('https://www.inpex.co.jp/')
```

```
import sqlite3

class dbar(object):
  def __init__(self, db_path):
    self.db = db_path
    self.table = 'urllist'
    self.conn = sqlite3.connect( db_path )
    self.c = self.conn.cursor()
    self.create_table()

  def create_table(self):
    self.c.execute('''
      create table if not exists {} (
        url text,
        src text,
        content_type  text,
        original int,
        gzip int,
        brotli int,
        gzip_srvd int,
        br_srvd int
      )
      '''.format( self.table ))
    self.conn.commit()
  
  def insert(self, url, src, content_type, origin_size, gzip_size, brotli_size, gzip_srvd=0, br_srvd=0):
    self.c.execute('''
      insert into {} values ("{}", "{}", "{}", {}, {}, {}, {}, {})
      '''.format(self.table, url, src, content_type, origin_size, gzip_size, brotli_size, gzip_srvd, br_srvd))
    self.conn.commit()

  def show(self):
    for r in self.c.execute('''select * from {}'''.format(self.table) ):
      print(r)

#db = dbar('/content/drive/My Drive/kitaPro/20200520_TechWeek/bar.db')
#db.insert('a', 'b', 'c', 0, 1, 3)
#db.show()


```

```
import pandas as pd

def do():
  #0 db init
  db = dbar('/content/drive/My Drive/kitaPro/20200520_TechWeek/bar.db')

  #1 get df225
  df225 = pd.read_csv('/content/drive/My Drive/kitaPro/20200520_TechWeek/df225.csv')

  #2 for loop in each website
  for url in df225['web'].head(1):
    ## 2.1 check base html + insert
    ct, orig, gzip, brotli, gzip_srvd, br_srvd = check_comp(url)
    db.insert(url, url, ct, orig, gzip, brotli, gzip_srvd, br_srvd)

    ## 2.2 get sub urls from base html
    urllist = suburls(url)
    ## 2.3 for loop: check each object + insert
    for u in urllist:
      ct, orig, gzip, brotli, gzip_srvd, br_srvd = check_comp(u)
      db.insert(u, url, ct, orig, gzip, brotli, gzip_srvd, br_srvd)

  #3. show tables
  db.show()

do()
```
